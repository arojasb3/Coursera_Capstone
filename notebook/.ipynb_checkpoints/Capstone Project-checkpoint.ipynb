## Introduction: Business Problem <a name="introduction"></a>

Imagine we want to create an app that is able to recomend neighborhoods based on specific requests: Population, employers, type of venues? For example, someone who wants to start a new business, maybe a new italian restaurant?: High population, low amount of restaurants so less competition and a decent amount of parks to get customers when they go out for a walk. 

Currently, there are available different apps to check specific recomendatios about places at the neighborhoods of a city (nextdoor, google maps, even foursquare). But for users who are moving into new cities, where they have zero to none clue about the new places they will be living at for the next few months or even years, we want to bring a solution to them, which will help in the process of selecting a neighborhood or at least to get a list of a couple neighborhoods which follow some of their requirements. It doesn't matter if they want it to start a business, live, invest, or whatever idea they may have.

## Methodology <a name="methodology"></a>
[Go Back to Table of Contents](#table_contents)

This recommender system will start as an mvp, using the algorithm teached by the IBM course "Machine Learning with Python" on the content based recommender systems.

First, we will have an initial matrix $M_{(nxv)}$ where every row represents a neighborhood and each column contains the proportion of the venue category for that neighborhood, $n$ is the number of neighborhoods and $v$ the number of venue categories. For example we can have "toronto neighborhood one" where it has 0.6 for restaurants, 0.2 for parks and 0 on the rest of the venue categories, of course this is only an example.

Now, our user will create an input with the desired venue values which is a vector $I_{1xk}$ where $k /leq v$, the venue categories doesn't need to be on the same order as the on the matrix $M$. Now we will organize the input for it to match the order of the venue categories from our data matrix, and create a new vector $\hat{I}_{1xv}$ where every $I_{i} \leq 10 \quad \forall i \leq v$. 10 will indicate that the user has great interest on that category and 0 that there is a low interest.

Then, our final recommendation $R$ will be calculated as 

$R = \sum_{i=1}^n \frac{(M \circ \hat{I})_i}{|\hat{I}|}$

where $\circ$ represents the element wise multiplication, we sum all the column values and divide by the total ratings of our user, leaving us with a $nx1$ vector with a value for each neighborhood. Then we will show the top 3 results in our app, and some of the extra information we gathered up on the Data section (crime rates, population).

For the data analysis we will plot histograms of the frequency of each venue category and try a k means clustering to see if our neighborhoods are pretty alike between them or not.

We can see that the most common venues are Stores, Restaurants, Fast Food, Cafes, even Parks. So looking for venues different than this ones may lead to a group of specific neighborhoods. Let's do a simple clustering algorithm in case we have neighborhoods in common. We will follow the same procedure done on the last notebook, I'm curious about the result with the new venue category groups

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]

num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# remove population variables from the grouped dataframe
toronto_grouped_venues = toronto_grouped.drop(['Total population', 'number of 15-45', 'number of employers'], 1)
# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped_venues['Neighborhood']

for ind in np.arange(toronto_grouped_venues.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped_venues.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

There you go, this is just for the clustering analysis at the end. We will try 4 clusters, just to check how the results are

```python
from sklearn.cluster import KMeans
import folium
from geopy.geocoders import Nominatim

address = 'Toronto, Ontario'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
```

```python
# set number of clusters
kclusters = 4

toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)
```

```python
# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

toronto_merged = df_toronto

toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)
toronto_merged.dropna(inplace=True)
# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighborhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[int(cluster)-1],
        fill=True,
        fill_color=rainbow[int(cluster)-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```

Cool, they are all mixed out. We could use some information about this clusters at the final output of our reccomender system

## Results <a name="results"></a>
[Go Back to Table of Contents](#table_contents)




Let's find out how our recommender system behaves.

For out input, we would desire our user to type a value between 1 and 10 depending on how important is that type of venue to our user. For example, Arts:10, Bank:1, Restaurant:6, Sports:8.

Let's create a simply user input situation:

```python
userInput = [
            {'category':'Arts', 'rating': 10},
            {'category':'Bank', 'rating': 1},
            {'category':'Restaurant', 'rating': 6},
            {'category':'Sports', 'rating': 8}
         ] 
input_user = pd.DataFrame.from_dict(userInput)
input_user
```

Now, we want to create the rating value of our user from all the avaiable categories, It has to be on the same order as in our data frame. 

```python
all_categories = pd.DataFrame(toronto_grouped.columns[1:])
all_categories.columns = ['category']
ratings = pd.merge(input_user, 
         all_categories, 
         left_on = 'category', right_on = 'category', how = 'right').fillna(0)
ratings = ratings.set_index('category')
ratings
```

Now we need to get the data from all of our neighborhoods! Obviously without the actual neighborhood name since it doesn't help for our recommendation system

```python
score_table = toronto_grouped.drop('Neighborhood', 1)
score_table.head()
```

The time has come, let's find the dot product between our user's preferences and the neighborhood dataframe. This should give us an score for each neighborhood, then we divide it by the total of the user preferences, weighing the ratings and puting them between 0 and 1. We will return the top 3 ones.

```python
recommendation_table_df = pd.DataFrame(np.dot(score_table,ratings.to_numpy()).sum(axis=1)/ratings.to_numpy().sum())
```

```python
recommendation_table_df
```

```python
toronto_grouped.shape
```

Good, our shapes match so we are sure we have a result for each neighborhood. Let's add the name, sort them and see our top recommended neighborhoods! =)

```python
recommendation_table_df['Neighborhood'] = toronto_grouped['Neighborhood']
recommendation_table_df.columns = ['score', 'Neighborhood']
recommendation_results = recommendation_table_df.sort_values(by='score',ascending=False).head(3)
recommendation_results
```

```python
recommendation_results['score'] = recommendation_results['score'].rank(ascending=False, method='max')
final_result = pd.merge(recommendation_results, 
         df_toronto, 
         left_on = 'Neighborhood', right_on = 'Neighborhood', how = 'inner')
final_result.pop('Neighbourhood')
final_result
```

Now that we have our top recommended neighborhoods, let's plot them in the map and hopefully we can help our users to make a final decision!

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)
#toronto_merged.dropna(inplace=True)
# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(final_result["Neighborhood"].size)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

color_dict = {}
for i in range(final_result["Neighborhood"].size):
    color_dict[final_result["Neighborhood"][i]] = i

# add markers to the map
markers_colors = []
for lat, lon, poi, score in zip(final_result['Latitude'], final_result['Longitude'], 
                         final_result['Neighborhood'], final_result['score']):
    label = folium.Popup(str(poi) + ' Score: ' + str(score), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[color_dict[poi]],
        fill=True,
        fill_color=rainbow[color_dict[poi]],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```

We can see the 3 different results plotted in the map, 1 neighbor is on the east part of the city while the other 2 are really close between them and are located around the middle west of the city. This gives an overall idea of the zones that our user may find suitable to live or start a new business.

We will now create a python function to test some other user inputs and check our reccomendations.

```python
def generate_recommendation(user_input, toronto_grouped, df_toronto, top_n=3):
    
    #process the user input
    input_user = pd.DataFrame.from_dict(user_input)
    all_categories = pd.DataFrame(toronto_grouped.columns[1:])
    all_categories.columns = ['category']
    
    # get the ratings of what the user desires
    ratings = pd.merge(input_user, 
             all_categories, 
             left_on = 'category', right_on = 'category', how = 'right').fillna(0)
    ratings = ratings.set_index('category')
    
    # generate score table on each neighborhood based on the user ratings
    score_table = toronto_grouped.drop('Neighborhood', 1)
    
    # generate final recommendation table
    recommendation_table_df = pd.DataFrame(
        np.dot(score_table,ratings.to_numpy()).sum(axis=1)/ratings.to_numpy().sum())
    recommendation_table_df['Neighborhood'] = toronto_grouped['Neighborhood']
    recommendation_table_df.columns = ['score', 'Neighborhood']
    
    # get top n results
    recommendation_results = recommendation_table_df.sort_values(by='score',ascending=False).head(top_n)
    recommendation_results['score'] = recommendation_results['score'].rank(ascending=False, method='max')
    
    # final results
    final_result = pd.merge(recommendation_results, 
             df_toronto, 
             left_on = 'Neighborhood', right_on = 'Neighborhood', how = 'inner')
    
    # clean df after merge (duplicate columns)
    final_result.pop('Neighbourhood')
    return final_result

def generate_map(latitude, longitude, final_result):
    # create map
    map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)
    #toronto_merged.dropna(inplace=True)
    # set color scheme for the clusters
    x = np.arange(final_result["Neighborhood"].size)
    ys = [i + x + (i*x)**2 for i in range(final_result["Neighborhood"].size)]
    colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
    rainbow = [colors.rgb2hex(i) for i in colors_array]

    color_dict = {}
    for i in range(final_result["Neighborhood"].size):
        color_dict[final_result["Neighborhood"][i]] = i

    # add markers to the map
    markers_colors = []
    for lat, lon, poi, score in zip(final_result['Latitude'], final_result['Longitude'], 
                             final_result['Neighborhood'], final_result['score']):
        label = folium.Popup(str(poi) + ' Score: ' + str(score), parse_html=True)
        folium.CircleMarker(
            [lat, lon],
            radius=5,
            popup=label,
            color=rainbow[color_dict[poi]],
            fill=True,
            fill_color=rainbow[color_dict[poi]],
            fill_opacity=0.7).add_to(map_clusters)

    return map_clusters
```

```python
address = 'Toronto, Ontario'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
```

```python
user_input = [
        {'category':'Arts', 'rating': 10},
        {'category':'Bank', 'rating': 1},
        {'category':'Restaurant', 'rating': 6},
        {'category':'Sports', 'rating': 8}
     ] 

result1 = generate_recommendation(user_input, deepcopy(toronto_grouped), deepcopy(df_toronto))
generate_map(latitude, longitude, result1)
```

```python
user_input = [
        {'category':'Bar', 'rating': 10},
        {'category':'Store', 'rating': 10},
        {'category':'Restaurant', 'rating': 10}
     ] 

result3 = generate_recommendation(user_input, deepcopy(toronto_grouped), deepcopy(df_toronto))
generate_map(latitude, longitude, result3)
```
